```python
!pip install --upgrade pip
!pip install pandas faker
from faker import Faker
import pandas as pd
import random

```

```python
# Inicializa o Faker
fake = Faker('pt_BR')
```

```python
# Definindo os gêneros, estados civis possíveis e fontes de tráfego
genders = ['Masculino', 'Feminino', 'Não-binário', 'Outro', 'Prefiro não dizer']
marital_statuses = ['Solteiro(a)', 'Casado(a)', 'Divorciado(a)', 'Viúvo(a)']
fonte_trafego = ['Orgânico', 'PPC', 'Redes Sociais', 'E-mail Marketing', 'Referência', 'Direto']
periodo=['Matutino','Vespertino','Noturno']
```

```python
# Lista de estados e cidades brasileiras (exemplo simplificado)
states_cities = {
    'São Paulo': ['São Paulo', 'Campinas', 'Santos'],
    'Rio de Janeiro': ['Rio de Janeiro', 'Niterói', 'Petrópolis'],
    'Minas Gerais': ['Belo Horizonte', 'Uberlândia', 'Contagem'],
    'Bahia': ['Salvador', 'Feira de Santana', 'Vitória da Conquista'],
    # Adicione mais estados e cidades conforme necessário
}
```

```python
# Gerar os dados
data = []
for _ in range(15000):
    estado = random.choice(list(states_cities.keys()))
    cidade = random.choice(states_cities[estado])
    nome = fake.first_name()
    sobrenome = fake.last_name()
    idade = random.randint(18, 80)
    genero = random.choice(genders)
    estado_civil = random.choice(marital_statuses)
    quantidade_de_filhos = random.randint(0, 5)
    renda_anual = round(random.uniform(12000, 200000), 2)  # Renda anual entre 12k e 200k
    fonte_de_trafego = random.choice(fonte_trafego)
    periodos= random.choice(periodo)

    data.append([
        nome,
        sobrenome,
        idade,
        genero,
        estado_civil,
        quantidade_de_filhos,
        renda_anual,
        estado,
        cidade,
        fonte_de_trafego,
        periodos
    ])
```

```python
# Criar um DataFrame
df = pd.DataFrame(data, columns=[
    'Nome',
    'Sobrenome',
    'Idade',
    'Gênero',
    'Estado Civil',
    'Quantidade de Filhos',
    'Renda Anual',
    'Estado',
    'Cidade',
    'Fonte_de_Trafego',
    'periodos'
])
```

```python
df.to_csv('base_segmentacao_clientes.csv', index=False)
```
